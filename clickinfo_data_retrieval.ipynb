```python
import pickle
import numpy as np
from datetime import date
import matplotlib.pyplot as plt
#import plotly.express as px

%matplotlib inline

pickle_in = open('left.pickle','rb')
left_data = pickle.load(pickle_in)

pickle_in = open('right.pickle','rb')
right_data = pickle.load(pickle_in)

# Normalizing the data for easy plotting
# actual screen resolution is 1920x1080 but it is scaled by 125%
# so (1920/1.25,1080/1.25) should be the actual screen area i.e,(1536,864)

click_left=[]
for p in left_data[str(date.today())]:
    a=[]
    a=p[0],p[1]
    click_left.append(a)
    

click_right=[]
for p in right_data[str(date.today())]:
    a=[]
    a=p[0],p[1]
    click_right.append(a)

left_x = []
for i in click_left:
    left_x.append(i[0])

left_y = []
for i in click_left:
    left_y.append(i[1])

right_x = []
for i in click_right:
    right_x.append(i[0])

right_y = []
for i in click_right:
    right_y.append(i[1])
    
# plt.scatter(left_x, left_y, color='r')
# plt.scatter(right_x,right_y, color='b')
# plt.xlabel('Horizontal view')
# plt.ylabel('Vertical view')
# plt.show()

fig1 = plt.figure()
l = plt.scatter(left_x,left_y,color='r')
r = plt.scatter(right_x,right_y, color='b')
plt.xlabel('Horizontal view')
plt.ylabel('Vertical view')
plt.legend((l,r), ('left_click', 'right_click'))

fig2 = plt.figure()
plt.hist2d(left_x, left_y, bins=10)
plt.xlabel('Horizontal view')
plt.ylabel('Vertical view')
cbar = plt.colorbar()
cbar.ax.set_ylabel(' Left clicks')

fig3 = plt.figure()
plt.hist2d(right_x, right_y, bins=10)
plt.xlabel('Horizontal view')
plt.ylabel('Vertical view')
cbar = plt.colorbar()
cbar.ax.set_ylabel(' Right clicks')
```
